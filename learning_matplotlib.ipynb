```python
#importing all neccasary modules
import matplotlib.pyplot as plt
#Pyplot is a module of Matplotlib which provides simple functions to add plot elements like lines, images, text, etc.
import numpy as np
```

```python
#Simple 2D plot of some coordinates
plt.plot([1,2,3,4],[2,4,9,12]) # here first array represant x axis and second array represent y axis of plot
```

```python
# we can add title, label for x axis and y axis in our plot
plt.plot([1,2,3,4],[2,4,9,12])
plt.title("Our learning plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#We can also specify the size of the figure using method figure() 
#and passing the values as a tuple of the length of rows and columns to the argument figsize
plt.figure(figsize=(10, 5))
plt.plot([1,2,3,4],[2,4,9,12])
plt.title("Our learning plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
# note that we have to use figure method before anything else for out plot
```

```python
#With every X and Y argument, you can also pass an optional third argument in the form of a string 
#which indicates the colour and line type of the plot. 
#The default format is b- which means a solid blue line. In the figure below we use ro which means red circles.
plt.figure(figsize=(10, 5))
plt.plot([1,2,3,4],[2,4,9,12],"ro")
plt.title("Our learning plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#plotting multiple sets of data in the same plot
x = np.arange(5)
y = x**3
plt.figure(figsize=(10, 5))
plt.plot([1,2,3,4],[2,4,9,12],"ro",x, y, "g^") # here we give second set of x and y values using comma operator
plt.title("Our learning plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# we can show multiple plots in one figure using subplot method
# subplot() takes three arguments (nrows, ncols, index) as in tuple
plt.subplot(1, 2, 1)
plt.plot([1,2,3,4], [1,4,9,16], "g^")
plt.title("first plot")

plt.subplot(1,2,2)
plt.plot(x,y,"ro")
plt.title("second plot")

plt.suptitle("2 plots in one figure")
plt.show()
```

```python
# we can also show plots in 2 rows and one coloums
plt.subplot(2, 1, 1)
plt.plot([1,2,3,4,5], [1,4,9,16,25], "g^")
plt.title("first plot")

plt.subplot(2,1,2)
x = np.arange(1, 6)
y = x**3
plt.plot(x,y,"ro")
plt.title("second plot")

plt.suptitle("2 plots in one figure")
plt.show()
```

```python
x = np.arange(1,6)
y = x**5
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax[0,0].plot(x ,y, "go")
ax[1,1].plot([1,2,3,4],[5,6,7,9], "r^")
ax[0,0].set_title("green_circles")
ax[1,1].set_title("red_triangle")
plt.show()
```

```python
# Creating bar graphs with matplotlib
# pyplot provides a method bar() to make bar graphs 
# which take arguments: categorical variables, their values and color.
cat_v = ["class_A", "class_B", "class_C","class_D", "class_E"]
val = [20, 60, 80, 40, 90]
plt.title("bar graph of marks")
plt.bar(cat_v, val, color="blue")
plt.xlabel("classes")
plt.ylabel("Marks")
```

```python
# creating horizontal bar graph with barh
cat_v = ["class_A", "class_B", "class_C","class_D", "class_E"]
val = [20, 60, 80, 40, 90]
plt.title("Horizontal bar graph of marks")
plt.barh(cat_v, val, color="green")
plt.xlabel("marks")
plt.ylabel("classes")
plt.show()
```

<b>To create horizontally stacked bar graphs we use the bar() method twice and pass the arguments where we mention the index and width of our bar graphs in order to horizontally stack them together. Also, notice the use of two other methods legend() which is used to show the legend of the graph and xticks() to label our x-axis based on the position of our bars.</b>

```python
cat_v = ["class_A", "class_B", "class_C","class_D", "class_E"]
avg_val = [20, 60, 80, 40, 90]
avg_boys = [21, 50, 70, 35, 60]

index = np.arange(5)
width = 0.3

plt.bar(index, avg_val, width, color="blue", label="average marks")
plt.bar(index+width, avg_boys, width, color="red", label="boys average marks")

plt.title("Horizontly slacked bar graph of marks")
plt.xlabel("classes")
plt.ylabel("Marks")
plt.xticks(index+width/2, cat_v)
plt.legend(loc="best")
plt.show()
```

```python
# To vertically stack the bar graphs together 
# we can use an argument bottom and mention the bar graph which we want to stack below as its value.
cat_v = ["class_A", "class_B", "class_C","class_D", "class_E"]
avg_girls_marks = [20, 60, 80, 40, 90]
avg_boys_marks = [21, 50, 70, 35, 60]

index = np.arange(5)
width = 0.3

plt.bar(index, avg_girls_marks, width, color="blue", label="girls average marks")
plt.bar(index, avg_boys_marks, width, color="red", label="boys average marks", bottom = avg_girls_marks)

plt.title("vertically slacked bar graph of marks")
plt.xlabel("classes")
plt.ylabel("Marks")
plt.xticks(index, cat_v)
plt.legend(loc="best")
plt.show()
```

<b>One more basic type of chart is a Pie chart which can be made using the method pie() We can also pass in arguments to customize our Pie chart to show shadow, explode a part of it, tilt it at an angle as follows:</b>

```python
subjects = ["sub_1", "sub_2", "sub_3", "sub_4", "sub_5", "sub_6"]
marks = [40, 45, 20, 60, 10, 95]
explode =[0, 0, 0, 0, 0, 0.1]
plt.figure(figsize=(10,10))
plt.pie(marks, explode=explode, labels = subjects, shadow = True, startangle = 30)
plt.axis('equal')
plt.legend(title="Marks in subjects")
plt.show()
```

<h3> Scatter plots and 3-D plotting </h3>

```python
height = np.random.randint(100, 500, 40)
weight = np.random.randint(10, 300, 40)
plt.xlim(100, 500)
plt.ylim(10, 300)
plt.scatter(height, weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("weight")
plt.show()
```

The above scatter can also be visualized in three dimensions. To use this functionality, we first import the module mplot3d as follows:

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(height, weight)
ax.set_xlabel("Height")
ax.set_ylabel("weight")
plt.show()
```

<h2>We can also create 3-D graphs of other types like line graph, surface, wireframes, contours, etc. The above example in the form of a simple line graph is as follows: Here instead of scatter3D() we use method plot3D()</h2>

```python
ax = plt.axes(projection='3d')
ax.plot3D(height, weight)
ax.set_xlabel("Height")
ax.set_ylabel("weight")
plt.show()
```
